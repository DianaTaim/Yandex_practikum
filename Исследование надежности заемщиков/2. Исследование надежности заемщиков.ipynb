## Предобработка данных

**Проверим количество пропущенных значений**

```python
print(data.isna().sum())
```

В двух столбцах есть пропущенные значения. Один из них — days_employed. Пропуски в этом столбце вы обработаете на следующем этапе. Другой столбец с пропущенными значениями — total_income — хранит данные о доходах. На сумму дохода сильнее всего влияет тип занятости, поэтому заполнить пропуски в этом столбце нужно медианным значением по каждому типу из столбца income_type

```python
for d in data['income_type'].unique(): 
    data.loc[(data['income_type']==d) & (data['total_income'].isna()), 'total_income'] = \
    data.loc[(data['income_type']==d), 'total_income'].median()
display(data.head())
```

**Обработаем аномальные значения в столбце days_employed, применим метод abs()**

```python
data['days_employed'] = data['days_employed'].abs()
data
```

**Выведем медианное значение для каждого типа занятости**

```python
days_employed_median = data.groupby('income_type')['days_employed'].median()
days_employed_median
```

У двух типов (безработные и пенсионеры) получатся аномально большие значения. Исправить такие значения сложно, поэтому оставим их как есть.

**Выведем уникальные значения столбца children**

```python
print(data['children'].unique())
```

В столбце children есть два аномальных значения. Удалим их

```python
data = data.loc[(data['children']!=-1)&(data['children']!=20)]
print(data['children'].unique())
```

**Заполним пропуски в столбце days_employed медианными значениями по каждому типу занятости income_type**

```python
for d in data['income_type'].unique():
    data.loc[(data['income_type'] == d)&(data['days_employed'].isna()),'days_employed']=\
    data.loc[(data['income_type'] == d), 'days_employed'].median()
data.head(10)
```

**Проверим пропущенные значения**

```python
data.isna().sum()
```

**Изменим вещественный тип данных в столбце total_income на целочисленный с помощью метода astype().**

```python
data['total_income']=data['total_income'].astype('int')
```

**Обработаем неявные дубликаты в столбце education**

```python
data['education']=data['education'].str.lower()
```

```python
display(data.duplicated().sum())
data=data.drop_duplicates()
display(data.head())
display(data.duplicated().sum())
```

**Создадим total_income_category с категориями:**
* 0–30000 — 'E';
* 30001–50000 — 'D';
* 50001–200000 — 'C';
* 200001–1000000 — 'B';
* 1000001 и выше — 'A'

Создадим функцию:

```python
def categorize_income(total_income):
    if total_income<=30000:
        return 'E'
    if total_income<=50000:
        return 'D'
    if total_income<=200000:
        return 'C'
    if total_income<=1000000:
        return 'B'
    return 'A'
```

```python
data['total_income_category'] = data['total_income'].apply(categorize_income)
```

Выведем перечень уникальных значений методом unique()

```python
data['purpose'].unique()
```

Создадим функцию, которая на основании данных из столбца purpose сформирует новый столбец purpose_category, в который войдут следующие категории:
* 'операции с автомобилем',
* 'операции с недвижимостью',
* 'проведение свадьбы',
* 'получение образования'.

```python
def categorize_purpose(index):
    if index == 'приобретение автомобиля' or index =='на покупку подержанного автомобиля' or index =='на покупку своего автомобиля' or index =='автомобили' or index =='сделка с подержанным автомобилем' or index =='автомобиль' or index =='свой автомобиль' or index =='сделка с автомобилем' or index =='на покупку автомобиля':
        return 'операции с автомобилем'
    if index == 'покупка жилья' or index =='операции с жильем' or index =='покупка жилья для семьи' or index =='покупка недвижимости' or index =='покупка коммерческой недвижимости' or index =='покупка жилой недвижимости' or index =='строительство собственной недвижимости' or index =='недвижимость' or index =='строительство недвижимости' or index =='операции с коммерческой недвижимостью' or index =='строительство жилой недвижимости' or index =='жилье' or index =='операции со своей недвижимостью' or index =='покупка своего жилья' or index =='операции с недвижимостью' or index =='покупка жилья для сдачи' or index =='ремонт жилью':
        return 'операции с недвижимостью'
    if index == 'сыграть свадьбу' or index =='на проведение свадьбы' or index =='свадьба':
        return 'проведение свадьбы'
    return 'получение образования'
```

```python
data['purpose_category']=data['purpose'].apply(categorize_purpose)
display(data)
```

## Исследовательский анализ

**Изучим есть ли зависимость между количеством детей и возвратом кредита в срок**

```python
stat=data.pivot_table(index=['children'], values='debt', aggfunc={'sum','count','mean'})
stat['mean']=stat['mean']*100
stat.reset_index()
```

**Вывод:** Клиенты, не имеющие детей, лучше покрывают свои крелитные обязательства 7,54%(только 1063 из 14091 имели задолженности по кредитам) в сравнении с семьями с 1-2 детьми это 9,23% и 9,45%.

**Выясним есть ли зависимость между семейным положением и возвратом кредита в срок**

```python
stat_by_status = data.pivot_table(index=['family_status'], values = 'debt', aggfunc={'count','sum','mean'})
stat_by_status['mean'] = stat_by_status['mean']*100
stat_by_status.reset_index()
```

**Вывод:** Лучше погашают свои кредитные обязательства клиенты из категории в женат/замужем - 7,56%(при количестве клиентов 12261 только у 927 были задолженности по кредитам). Хуже всего отдают кредиты клиенты в категориях граждаснкий брак (9,31%) и не женат/не замужем (9,76%)

**Выясним есть ли зависимость между уровнем дохода и возвратом кредита в срок**

```python
data_by_income = data.pivot_table(index=['total_income_category'], values='debt', aggfunc={'sum','count','mean'})
data_by_income['mean']=data_by_income['mean']*100
data_by_income.reset_index()
```

**Вывод:** Кредитополучатели, зарабатывающие в диапазоне "В" лучше погашают свои кредитные обязательства(7,06%) против кредитополучателей в категории "С" - 8,49%. Разница между этими двумя категориями почти 1,5%

**Проверим как разные цели кредита влияют на его возврат в срок**

```python
data_by_purpose = data.pivot_table(index=['purpose_category'], values = 'debt', aggfunc={'sum', 'count', 'mean'})
data_by_purpose['mean']=data_by_purpose['mean']*100
data_by_purpose.reset_index()
```

**Вывод:** Лучше всего ситуация по задолженностям в категории "операции с недвижимостью" - 7,25% (при рассмотрении 10751 клиентов только у 780 были задолженности). Хуже всего ситуация обстоит по категориям "операции с автомобилем" и "получение образования" - это 9,34% и 9,25% соответственно. На мой взгляд самые надежные кредиторы те, кто берет ипотеку связан с постепенным удорожанием такого актива как недвижимость, ежегодная инфляция и др. причины, в то время как такой актив как автомобиль ежегодно теряет в своей стоимости + ежемесячные расходы на автомобиль(топливо, текущий ремонт, страховка и пр), а также возможные непредвиденные обстоятельства(аварии и пр)
